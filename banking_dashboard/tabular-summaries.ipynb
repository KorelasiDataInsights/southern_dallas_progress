# **Add Race_Ethnicity field to HDMA data** 

if the race is in (Asian, asian indian, other asian, vietnamese, filipino, Korean, Japanese) then asian 

if hmda_dict_lar_df['applicant_race_1'] == "White" and "Hispanic or Latino" in hmda_dict_lar_df['applicant_ethnicity_1'] then hmda_dict_lar_df['applicant_race_1'] +' '+ hmda_dict_lar_df['applicant_ethnicity_1']

if race == "Black or African American" then "Black or African American"

else "Other"

```python
def race_ethnicity_groups(race_eth_str:str):
    if race_eth_str in ['White']:
        return 'White'
    elif race_eth_str in ['Asian', 'Asian Indian', 'Other Asian', 'Vietnamese', 'Filipino', 'Korean', 'Japanese']:
        return 'Asian'
    elif race_eth_str in ['Black or African American']:
        return 'Black or African American'
    elif race_eth_str in ['Not applicable']:
        return 'Not applicable'
    elif race_eth_str in ['Information not provided by applicant in mail, internet, or telephone application']:
        return 'Unspecified'
    else:
        return 'Other'
```

```python
hmda_dict_lar_df['Race_Ethnicity'] = hmda_dict_lar_df['applicant_race_1'].apply(race_ethnicity_groups)
```

```python
# find rows in 'Race_Ethnicity' that meet specified conditions and concatenate race and ethnicity if conditions met otherwise return value that is 
# already present
hmda_dict_lar_df['Race_Ethnicity'] = np.where((hmda_dict_lar_df['applicant_race_1'] == "White") &\
                                     ((hmda_dict_lar_df['applicant_ethnicity_1'] == "Not Hispanic or Latino")|\
                                     (hmda_dict_lar_df['applicant_ethnicity_1'] == "Hispanic or Latino")|\
                                     (hmda_dict_lar_df['applicant_ethnicity_1'] == "Other Hispanic or Latino"))\
                                    , hmda_dict_lar_df['applicant_race_1'] +' '+ hmda_dict_lar_df['applicant_ethnicity_1']\
                                    , hmda_dict_lar_df['Race_Ethnicity'])
```

```python
ffiec_intermed_df = ffiec_data.groupby(['County','Census tract'], as_index = False)\
[['Total persons','Total families','Total households','Income indicator, which identifies low, moderate, middle, and upper income areas']].sum()
```

```python
# hmda originated
hmda_lei_count = hmda_dict_lar_df[hmda_dict_lar_df['action_taken'] == 'Loan originated'].groupby(['county_code','census_tract'], as_index = False)\
[['lei']].count().rename(columns = {'lei':'lei_count'})
```

```python
hmda_lei_count.head(10)
```

```python
hmda_race_prcnt_df = hmda_dict_lar_df[hmda_dict_lar_df['action_taken'] == 'Loan originated'].groupby(['county_code','census_tract'])\
[['Race_Ethnicity']].value_counts(normalize = True)
```

```python
hmda_race_prcnt_df = hmda_race_prcnt_df.unstack().reset_index()
```

```python
hmda_race_prcnt_df
```

```python
hmda_intermed_df = pd.merge(hmda_race_prcnt_df,hmda_lei_count, left_on = ['county_code','census_tract'],\
                            right_on = ['county_code','census_tract'], how = 'left')
```

```python
hmda_intermed_df
```

```python
sba_intermed_df = sba_data.groupby(['county_final','_tract'], as_index = False)[['Borrower identifier']].count()
```

```python
sba_intermed_df
```

```python
# merge hmda grouping to ffiec grouping
ffiec_hmda_intermed_df = pd.merge(ffiec_intermed_df, hmda_intermed_df, left_on = ['County', 'Census tract']\
                                  , right_on = ['county_code','census_tract'], how = 'left')
```

```python
ffiec_hmda_intermed_df
```

```python
# merge sba to ffiec hmda merge
ffiec_hmda_sba_intermed_df = pd.merge(ffiec_hmda_intermed_df, sba_intermed_df, left_on = ['County', 'Census tract'], \
                                      right_on = ['county_final','_tract'], how = 'left').rename(columns = {'Borrower identifier':'sba_loan_count'})
```
