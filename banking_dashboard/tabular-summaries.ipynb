```python
# define helper function to for data pre-processing 
def bin_category(df:pd.core.frame.DataFrame, cols_to_check:list, threshold:int, normalize_inpt:bool = False, cnt_or_prp = 'count'):
    for category_col in cols_to_check:
        temp_df = df[category_col].value_counts(normalize = normalize_inpt).reset_index()
        replace_map = {val_under_threshold:'Other' for val_under_threshold in temp_df[temp_df[cnt_or_prp] < threshold][category_col]}
        if len(replace_map) > 1:
            df[category_col] = df[category_col].replace(replace_map)
    return df
```

# **Add Race_Ethnicity field to HDMA data** 

if the race is in (Asian, asian indian, other asian, vietnamese, filipino, Korean, Japanese) then asian 

if hmda_dict_lar_df['applicant_race_1'] == "White" and "Hispanic or Latino" in hmda_dict_lar_df['applicant_ethnicity_1'] then hmda_dict_lar_df['applicant_race_1'] +' '+ hmda_dict_lar_df['applicant_ethnicity_1']

if race == "Black or African American" then "Black or African American"

else "Other"

```python
def race_ethnicity_groups(race_eth_str:str):
    if race_eth_str in ['White']:
        return 'White'
    elif race_eth_str in ['Asian', 'Asian Indian', 'Other Asian', 'Vietnamese', 'Filipino', 'Korean', 'Japanese']:
        return 'Asian'
    elif race_eth_str in ['Black or African American']:
        return 'Black or African American'
    elif race_eth_str in ['Not applicable']:
        return 'Not applicable'
    elif race_eth_str in ['Information not provided by applicant in mail, internet, or telephone application']:
        return 'Unspecified'
    else:
        return 'Other'
```

```python
hmda_dict_lar_df['Race_Ethnicity'] = hmda_dict_lar_df['applicant_race_1'].apply(race_ethnicity_groups)
```

```python
# find rows in 'Race_Ethnicity' that meet specified conditions and concatenate race and ethnicity if conditions met otherwise
# return value that is already present
hmda_dict_lar_df['Race_Ethnicity'] = np.where((hmda_dict_lar_df['applicant_race_1'] == "White") &\
                                     ((hmda_dict_lar_df['applicant_ethnicity_1'] == "Not Hispanic or Latino")|\
                                     (hmda_dict_lar_df['applicant_ethnicity_1'] == "Hispanic or Latino")|\
                                     (hmda_dict_lar_df['applicant_ethnicity_1'] == "Other Hispanic or Latino"))\
                                    , hmda_dict_lar_df['applicant_race_1'] +' '+ hmda_dict_lar_df['applicant_ethnicity_1']\
                                    , hmda_dict_lar_df['Race_Ethnicity'])
```

```python
# census_level_hmda_num_summaries

census_level_hmda_num_summaries = {}
for num_field in ['loan_amount','interest_rate','origination_charges']:
    # get five number summaries for hmda data 
    temp_hmda_df = hmda_dict_lar_df.copy()
    temp_hmda_df[num_field] = temp_hmda_df[num_field].replace('Exempt',np.nan).apply(float)
    temp_num_counts = temp_hmda_df.groupby(['county_code','census_tract'])[num_field].describe().reset_index()
    # bring in ffiec subset 
    tract_inc_ind = ffiec_data[['County','Census tract','Tract Income Indicator']]
    # left merge ffiec data with 5 number summary df 
    temp_num_summary = pd.merge(temp_num_counts,tract_inc_ind, left_on = ['county_code','census_tract'],\
                                 right_on = ['County','Census tract'], how = 'left')
    # add df to dictionary and reorder column names 
    census_level_hmda_num_summaries[num_field] =\
    temp_num_summary[['county_code','census_tract','Tract Income Indicator','count','min','50%','mean','max']]\
    .fillna(0).rename(columns = {'50%':'median'})
    
```
